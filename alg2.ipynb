```python
import pandas as pd
d=pd.read_csv('/content/water_potability.csv')
d=d.fillna(0)
print(d)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
print(d.isnull())
x=d.iloc[ :,0:9]
y=d.iloc[ :,9:10]
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=True)
```

```python
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
mm=lr.fit(x_train,y_train)
```

```python
yp=mm.predict(x_test)
```

```python
print(yp)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(yp,y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier(criterion='entropy',random_state=0)
mm=classifier.fit(x_train,y_train)
```

```python
yp=mm.predict(x_test)
```

```python
print(yp)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(yp,y_test))
```

```python
from sklearn import metrics
```

```python
confusion_matrix=metrics.confusion_matrix(y_test,yp)
```

```python
cm_display=metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix,display_labels=[False,True])
```

```python
import matplotlib.pyplot as plt
```

```python
cm_display.plot()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
classifier.fit(x_train,y_train)
```

```python
yp=classifier.predict(x_test)
```

```python
print(yp)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(yp,y_test))
```

```python
confusion_matrix=metrics.confusion_matrix(y_test,yp)
```

```python
cm_display=metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix,display_labels=[False,True])
```

```python
cm_display.plot()
plt.show()
```
